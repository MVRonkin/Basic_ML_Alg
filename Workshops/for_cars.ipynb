Считываем файл в DataFrame - для этого использууем  метод .read_csv(path,delimiter) 



```python
import pandas as pd
import numpy as np
df = pd.read_csv('../Notebooks_for_textbook/cars.csv', delimiter = ',')
df
```

посмотрим на содержание таблицы

```python
df=df.rename(columns={'Transmission':'AutoTransmission'})
df['AutoTransmission'] = df['AutoTransmission'].map({'Automatic': True, 'Manual': False})
df
```

```python
df.columns
```

```python
# Список колонок, в которых хотим создать пропуски
columns_to_modify = ['Make', 'Model','Engine_capacity(cm3)',
       'Fuel_type']

# Процент пропусков
percent_missing = 0.5 / 100  # Переводим проценты в долю (например, 0,5% => 0.005)

for col in columns_to_modify:
    num_missing = int(len(df) * percent_missing)  # Число пропусков
    
    # Генерируем случайные индексы
    missing_indices = np.random.choice(df.index, size=num_missing, replace=False)
    
    # Устанавливаем значения в указанных индексах равными NaN
    df.loc[missing_indices, col] = np.nan

# Проверяем наличие пропусков
missing_values_count = df.isna().sum()
print(missing_values_count)
```
